```python
#Matrix Conguential Generator
import numpy as np

def matrix_multiplicative_congruential_generator(A, x_0, m, N):
    k = A.shape[0]
    X = np.zeros((N, k))
    X[0] = x_0
    for i in range(1, N):
        X[i] = np.dot(A, X[i-1]) % m
    return X / m

# Example usage:
A = np.array([[1234, 5678], [91011, 1213]])  # Example matrix A
x_0 = np.array([10, 20])  # Initial state vector
m = 2**31 - 1  # Modulus
N = 10  # Number of samples

# Generate random numbers
random_numbers = matrix_multiplicative_congruential_generator(A, x_0, m, N)
print(random_numbers)
```

```python
### Gap Test
import numpy as np
import matplotlib.pyplot as plt

def LCG(a,c,x_0,m,N):
    X = np.zeros(N+1)
    U = np.zeros(N+1)
    X[0] = x_0
    for i in range(1,N+1):
        X[i] = (a*X[i-1] + c)%m
        U[i]= X[i]/m
    return  U[1:]
a = 7**5
m = 2**31 -1
c = 0
x_0 = 10
N = 1000
U = LCG(a,c,x_0,m,N)

## Gap Test
alpha = 0.4
beta = 0.6
r = 7
plt.plot(np.arange(1,N+1),U,'.')
plt.show()

T = [] # creating an empty python list
T.append(0)  # T_0 = 0

# Collecting values in (0.4,0.6)  
for i in range(N):
    if 0.4<U[i]<0.6:
        T.append(i)
T = np.array(T) # Converting the list to a numpy array
print("T: ",T)

Z = np.zeros(len(T)-1) 
for i in range(len(T)-1):
    Z[i] = T[i+1]-T[i]-1
print("Z: ",Z)

Z = Z.astype(int)
Z_classes = np.zeros(r+1)
for i in range(len(Z)):  # Calculating Z classes
        if Z[i]<r:
            Z_classes[Z[i]] += 1
        else:
            Z_classes[r] +=1
print("Z_classes",Z_classes)

# Obtaining the observed class proportions
Z_obs = Z_classes/Z_classes.sum()
print(Z_obs)

p = beta-alpha

# Calculating the expected values for the geometric distribution
Z_expected = np.zeros(r+1)
for i in range(r+1): 
    if i<r :
        Z_expected[i] = p*(1-p)**i
    else :
        Z_expected[i] = (1-p)**r
        
print("Expected Z: ",Z_expected)

# Calculating the Chi-sqaure Statistic
D = ((Z_obs-Z_expected)**2/Z_expected).sum()
print("D Value: ",D)
```

```python
### Geometric Distribution 
import numpy as np
import matplotlib.pyplot as plt

def LCG(a, c, x_0, m, N):
    X = np.zeros(N + 1)
    U = np.zeros(N + 1)
    X[0] = x_0
    for i in range(1, N + 1):
        X[i] = (a * X[i - 1] + c) % m
        U[i] = X[i] / m
    return U[1:]

U = LCG(7**5,0,10,2**31-1,1000)

# Geometric Distribution
def EXP(lamda):
    E = np.zeros(len(U))
    for i in range(len(U)):
        E[i] = (-1) * (1 / lamda) * (np.log(U[i]))
    return E

def GEOM(p):
    Y = EXP((-1) * np.log(1 - p))
    G = np.zeros(len(U))
    for i in range(len(U)):
        G[i] = np.ceil(Y[i])
    return G

Vals = GEOM(0.5)
print(Vals)

plt.hist(Vals,bins=20,edgecolor="black",color="purple")
plt.show()
```

```python
### Uniform (Discrete) Distribution
import numpy as np
import matplotlib.pyplot as plt

def LCG(a, c, x_0, m, N):
    X = np.zeros(N + 1)
    U = np.zeros(N + 1)
    X[0] = x_0
    for i in range(1, N + 1):
        X[i] = (a * X[i - 1] + c) % m
        U[i] = X[i] / m
    return U[1:]

a = 7 ** 5
m = 2 ** 31 - 1
c = 0
x_0 = 10
N = 1000

U = LCG(a, c, x_0, m, N)

## Uniform (Discrete) Distribution
DU = np.zeros(len(U))
def UDD(a,b):
    for i in range(len(U)):
        DU[i]=np.floor(a+(b+1-a)*U[i])
    return DU
UD = UDD(1,10)
print(UD)

plt.hist(UD, bins=10,color="red",edgecolor="black")
plt.show()
```

```python
### Beta Distribution
import numpy as np
import matplotlib.pyplot as plt

def LCG(a, c, x_0, m, N):
    X = np.zeros(N + 1)
    U = np.zeros(N + 1)
    X[0] = x_0
    for i in range(1, N + 1):
        X[i] = (a * X[i - 1] + c) % m
        U[i] = X[i] / m
    return U[1:]

U = LCG(7**5,0,100,2**31-1,1000)

def gamma_distribution(alpha, size):
    d = alpha - 1/3
    c = 1 / np.sqrt(9 * d)
    
    # Initialize array to store generated random numbers
    X = np.zeros(size)
    
    # Step 2: Generate random numbers
    for i in range(size):
        while True:
            # Generate Z from N(0,1) and U from U(0,1) independently
            Z = np.random.normal(0, 1)
            V = (1 + c * Z)**3
            
            # Condition for acceptance
            if Z > -1/c and np.log(U[i]) < 0.5 * Z**2 + d - d * V + d * np.log(V):
                # Step 3: Generate X
                X[i] = d * V
                break
    
    return X

Y1 = gamma_distribution(0.5,100)
Y2 = gamma_distribution(0.5,100)

X = Y1/(Y1+Y2)
print(X)
plt.plot(np.linspace(0, 1, 100),X,color="red")
plt.show()
```

```python
import numpy as np

def beta_distribution(alpha, beta, size=1):
    Y1 = np.random.gamma(alpha, 1, size)
    Y2 = np.random.gamma(beta, 1, size)
    return Y1 / (Y1 + Y2)

# Example usage:
alpha = 2
beta = 3
size = 1000
X = beta_distribution(alpha, beta, size)
print(X)
plt.plot(np.linspace(0, 1, 1000),X,color="red")
plt.show()
```

```python
### Cauchy Distribution
## Method 01
import numpy as np
import matplotlib.pyplot as plt

def LCG(a, c, x_0, m, N):
    X = np.zeros(N + 1)
    U = np.zeros(N + 1)
    X[0] = x_0
    for i in range(1, N + 1):
        X[i] = (a * X[i - 1] + c) % m
        U[i] = X[i] / m
    return U[1:]

a = 7 ** 5
m = 2 ** 31 - 1
c = 0
x_0 = 10
N = 1000

U = LCG(a, c, x_0, m, N)

## Cauchy Distribution
X = np.tan(np.pi*U -(np.pi/2))
print(X)

plt.hist(X,bins=20,edgecolor="black",color="blue")
plt.show()
```

```python
### Exponential Distribution
import numpy as np
import matplotlib.pyplot as plt

def LCG(a, c, x_0, m, N):
    X = np.zeros(N + 1)
    U = np.zeros(N + 1)
    X[0] = x_0
    for i in range(1, N + 1):
        X[i] = (a * X[i - 1] + c) % m
        U[i] = X[i] / m
    return U[1:]

a = 7 ** 5
m = 2 ** 31 - 1
c = 0
x_0 = 10
N = 1000

U = LCG(a, c, x_0, m, N)

## Exponential Distribution
E = np.zeros(len(U))
def EXP(lamda):
    for i in range(len(U)):
        E[i]=(-1)*(1/lamda)*(np.log(U[i]))
    return E

lamda = 1
Y = EXP(lamda)
print(Y)

plt.hist(Y,edgecolor="black",color="orange")
plt.show()
```
